#6. How do we handle categorical variables in Machine Learning? What are the common techniques
Ans)
To use categorical variables in ML models, we convert them into numerical form.

**Common Techniques:**

**Label Encoding**

Assigns each category a number.

Example: Red=0, Blue=1, Green=2.

**One-Hot Encoding**

Creates separate binary columns for each category.

Example: Red = [1,0,0], Blue = [0,1,0].

**Ordinal Encoding**

For ordered categories (low < medium < high).

#15. What is causation? Explain difference between correlation and causation with an example.

Ans)
**Causation:**
Causation means that one variable directly makes another variable change. It is a cause-and-effect relationship.

**Difference between correlation and causation (simple explanation):**

Correlation means two variables move together, but one does not necessarily affect the other.

Causation means one variable actually causes the change in another.

Example:
During summer, both ice cream sales increase and drowning cases increase.
These two things are correlated because they happen at the same time.
But eating ice cream does not cause drowning.

#16. What is an Optimizer? What are different types of optimizers? Explain each with an example.
Ans)
**Optimizer:**
An optimizer is an algorithm that updates the model's parameters (like weights) to minimize the loss function during training.

**Types of Optimizers: **

**Gradient Descent (GD)**

Updates parameters using the gradient of the loss function over the entire dataset.

Example: Linear Regression trained on all data at once.

**Stochastic Gradient Descent (SGD)**

Updates parameters using the gradient for one sample at a time.

Example: Neural network updated after each training example.

**Mini-batch Gradient Descent**

Updates parameters using small batches of data.

Example: Training CNN with batch size = 32.
